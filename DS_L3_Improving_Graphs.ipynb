# Barplot

---
## Basic Barplot
---

```
# http://python-graph-gallery.com/barplot/
# https://matplotlib.org/tutorials/introductory/pyplot.html 
# Make a fake dataset:
height = [3, 12, 5, 18, 45]
bars = ('A', 'B', 'C', 'D', 'E')

# Create bars
plt.bar(bars, height)
 
# Create names on the x-axis
plt.xticks(bars)
 
# Show graphic
plt.show()
```

---
## Horizontal Barplot
---

```
# Create horizontal bars
plt.barh(bars, height)
 
# Create names on the y-axis
plt.yticks(bars)
 
# Show graphic
plt.show()
```

---
## Adjusting the colors
---

```
# RGB is a way of making colors.
# color =  red, green and blue + the transparency
# Put transparency = 0 
# Put transparency = 1
plt.bar(bars, height, color=(0.2, 0.4, 0.6, 0.7))
plt.xticks(bars)
plt.show()
```

```
# Different color for each bar 
# Provide a list of color names to the color argument
plt.bar(bars, height, color=['black', 'red', 'green', 'blue', 'cyan'])
plt.xticks(bars)
plt.show()
```

```
# Control color of the border 
# The edgecolor argument allows to color the borders of barplots.
plt.bar(bars, height, color=(0.1, 0.1, 0.1, 0.3),  edgecolor='blue')
plt.xticks(bars)
plt.show()
```

---
## Add title and axis labels 
---

```
# Create bars and choose color

plt.bar(bars, height, color = (0.5,0.1,0.5,0.6))

# Add title and axis names
plt.title('$Height_{bars}$', fontsize=14)
plt.xlabel('Name', fontsize=14)
plt.ylabel('Height', fontsize=14)
 
# Limits for the Y axis
plt.ylim(0,60)
 
# Create names
plt.xticks(bars)
 
# Show graphic
plt.show()
```

# Line chart

---
## Basic Line Chart
---

```
# Create a dataframe 
df = pd.DataFrame({'year': [2015, 2016, 2017, 2018, 2019], 'T20': [24, 36, 12, 18, 25], 'ODI': [89, 24, 37, 63, 98]})

# Draw the plot 
plt.plot( 'year', 'T20', data = df)
plt.show()
```

```
# Draw the plot 
# https://matplotlib.org/3.2.1/api/_as_gen/matplotlib.pyplot.legend.html 
#plt.plot( 'year', 'T20','k--', label = 'T20 score', data = df, linewidth = 5)
plt.plot( 'year', 'T20','k', data = df, linewidth = 5)
plt.plot('year', 'ODI','k:', data = df, c='red', linewidth = 5)
plt.xticks(np.arange(min(df.year), max(df.year)+1, 1.0), fontsize = 15)
plt.yticks(np.arange(0,125,25), fontsize = 15)
plt.title('Points scored by India', fontsize = 15)
plt.xlabel('Year', fontsize = 15)
plt.ylabel('Points', fontsize = 15)
plt.legend(loc='upper center')
plt.grid(b = True, axis = 'y', alpha=0.9) #which argument to choose the axis axis = 'x'/'y'/'both'
plt.show()
```

##  Multiple lines chart
---

```
# Data
df=pd.DataFrame({'x': range(1,11), 'y1': np.random.randn(10), 'y2': np.random.randn(10)+range(1,11), 'y3': np.random.randn(10)+range(11,21) })
 
# multiple line plot
plt.plot( 'x', 'y1', data=df, marker='o', markerfacecolor='blue', markersize=12, color='skyblue', linewidth=4)
plt.plot( 'x', 'y2', data=df, marker='', color='olive', linewidth=2)
plt.plot( 'x', 'y3', data=df, marker='', color='olive', linewidth=2, linestyle='dashed', label="toto")
plt.legend(loc='upper center') 
```

# Scatter plot 

---
## Basic scatter plot 
---

```
# Loading the data from an URL
url = "https://raw.githubusercontent.com/shala2020/shala2020.github.io/master/Lecture_Materials/Google_Colab_Notebooks/DataScience/L2/moviesData.csv"
movies = pd.read_csv(url)
movies.head()
```

```
# use the function regplot to make a scatterplot with regression line 
# sns.set(color_codes=True) 
sns.regplot(x = movies["critics_score"], 
            y = movies["audience_score"], ci = 95) # ci = None will not display the confidence interval 

# Without regression fit:
# sns.regplot(x=movies["critics_score"], 
#             y = movies["audience_score"], 
#             fit_reg=False)
```

```
# Another example to show the confidence interval
# https://www.mathsisfun.com/data/confidence-interval.html
tips = sns.load_dataset("tips")
# tips.head()
sns.regplot(x="total_bill", y="tip", data=tips, ci = None)
# sns.regplot(x="total_bill", y="tip", data=tips)
# sns.regplot(x="total_bill", y="tip", data=tips, ci = 95)
```

---
## Customizing the scatter plot
---

```
sns.regplot(x = movies["critics_score"], 
            y = movies["audience_score"], 
            marker="+", fit_reg=False)
```

```
sns.regplot(x = movies["critics_score"], 
            y = movies["audience_score"], 
            marker="+", fit_reg=False, color='red')
```

```
sns.regplot(x = movies["critics_score"], 
            y = movies["audience_score"], 
            marker="+", fit_reg=False, scatter_kws={"color":"red", "alpha":0.5, "s":100})
```

---
## Using categorical variable for color 
---

```
# Use the 'hue' argument to provide a factor variable
sns.lmplot(x = "critics_score", 
            y = "audience_score", data = movies, fit_reg=False, hue='genre', legend=False)#regplot is a subset of lmplot
 
# Move the legend (labeling of plot) to an empty part of the plot
plt.legend(loc='lower right')
```

```
df = sns.load_dataset('iris')
 
# Use the 'hue' argument to provide a factor variable
sns.lmplot(x="sepal_length", y="sepal_width", data=df, fit_reg=False, hue='species', legend=False)
 
# Move the legend to an empty part of the plot
plt.legend(loc='lower right')
```

## Plots for visualizing distribution


---
### Violin plot 
---


```
# https://python-graph-gallery.com/violin-plot/
# Violinplots allow to visualize the distribution of a numeric variable for one or several groups. 
# It is really close from a boxplot, but allows a deeper understanding of the density. 
# Violins are particularly adapted when the amount of data is huge and showing individual observations gets impossible. 
# Violinplots probably deserve more attention compared to boxplot that can sometimes hide features of the data.

# If you have only one numerical variable, you probably better have to make an histogram or a density plot. 
# But you can still use the violinplot function to describe the distribution of this variable, as follow:
from pylab import *
df = sns.load_dataset('iris')
 
plot
sns.violinplot(x=df["species"], y=df["sepal_length"])
sns.boxplot(x="species", y="sepal_length", data=df)
```

---
## Density plot 
---

```
# plot of 2 variables
# Kernel Density Estimate is used for visualizing the Probability Density of a continuous variable
p1=sns.kdeplot(df['sepal_width'], shade = True, color="r")
p1=sns.kdeplot(df['sepal_length'], shade=True, color="b")
```
